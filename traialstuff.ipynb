```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('Image/tower.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"
```

```python
b, g, r = cv2.split(img)

b_energy_sobel = np.absolute(cv2.filter2D(b, -1, np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]]))
                                + cv2.filter2D(b, -1, np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])))
g_energy_sobel = np.absolute(cv2.filter2D(g, -1, np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]]))
                                + cv2.filter2D(g, -1, np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])))
r_energy_sobel = np.absolute(cv2.filter2D(r, -1, np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]]))
                                + cv2.filter2D(r, -1, np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])))

print(b_energy_sobel.shape)

sobel = b_energy_sobel + g_energy_sobel + r_energy_sobel

plt.imshow(sobel)
plt.axis('off')
plt.show()
```

```python
b,g,r = cv2.split(img)

b_energy_scharr = np.absolute(cv2.filter2D(b, -1, np.array([[-3, 0, 3], [-10, 0, 10], [-3, 0, 3]]))
                                + cv2.filter2D(b, -1, np.array([[-3, -10, -3], [0, 0, 0], [3, 10, 3]])))
g_energy_scharr = np.absolute(cv2.filter2D(g, -1, np.array([[-3, 0, 3], [-10, 0, 10], [-3, 0, 3]]))
                                + cv2.filter2D(g, -1, np.array([[-3, -10, -3], [0, 0, 0], [3, 10, 3]])))
r_energy_scharr = np.absolute(cv2.filter2D(r, -1, np.array([[-3, 0, 3], [-10, 0, 10], [-3, 0, 3]]))
                                + cv2.filter2D(r, -1, np.array([[-3, -10, -3], [0, 0, 0], [3, 10, 3]])))


scharr = b_energy_scharr+ g_energy_scharr+ r_energy_scharr

# plot
plt.imshow(scharr)
plt.axis('off')
plt.show()
```

```python
b, g, r = cv2.split(img)

b = cv2.convertScaleAbs(b)
g = cv2.convertScaleAbs(g)
r = cv2.convertScaleAbs(r)


b_energy_canny = cv2.Canny(b, 100, 200)
g_energy_canny = cv2.Canny(g, 100, 200)
r_energy_canny = cv2.Canny(r, 100, 200)

print(b_energy_canny.shape)

canny = b_energy_canny+ g_energy_canny+ r_energy_canny

plt.imshow(canny)
plt.axis('off')
plt.show()
```

```python
b, g, r = cv2.split(img)

b = cv2.convertScaleAbs(b)
g = cv2.convertScaleAbs(g)
r = cv2.convertScaleAbs(r)

b_energy_prewitt = np.absolute(cv2.filter2D(b, -1, kernel=np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))) + \
                    np.absolute(cv2.filter2D(b, -1, kernel=np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])))
g_energy_prewitt = np.absolute(cv2.filter2D(g, -1, kernel=np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))) + \
                    np.absolute(cv2.filter2D(g, -1, kernel=np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])))
r_energy_prewitt = np.absolute(cv2.filter2D(r, -1, kernel=np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))) + \
                    np.absolute(cv2.filter2D(r, -1, kernel=np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])))
                            
prewitt = b_energy_prewitt + g_energy_prewitt + r_energy_prewitt

# show only the image and dont show the axis
plt.imshow(prewitt)
plt.axis('off')
plt.show()
```
